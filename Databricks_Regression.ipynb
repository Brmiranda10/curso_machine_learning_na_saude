```
import pandas as pd
!wget http://localhost/mimic_proc_f_test.csv.gz
mimic = pd.read_csv('mimic_proc_f_test.csv.gz') # 5.565
# mimic = pd.read_csv('mimic_proc_m_test.csv.gz') # 7.323
# mimic = pd.read_csv('mimic_proc_f_train.csv.gz') # 16.693
# mimic = pd.read_csv('mimic_proc_m_train.csv.gz') # 21.968
len(mimic), mimic.columns
```

```
columns_data = mimic.columns[1:len(mimic.columns)-4]
data = mimic[columns_data].fillna(0).values
target = mimic['cC#stay'].values
print(len(target))
```

```
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

train_data, test_data, train_target, test_target = train_test_split(data, target, test_size=0.33)
print(len(train_data))

regressor = RandomForestRegressor()

regressor.fit(train_data, train_target)
```

```
regressor.predict(test_data)
```

```
from sklearn.metrics import mean_absolute_error

prediction = regressor.predict(test_data)
truth = test_target

mean_absolute_error(truth, prediction)
```

```
from sklearn.model_selection import cross_val_score

cross_val_score(regressor, data, target, scoring='neg_mean_absolute_error')
```

```
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8), dpi= 300)
plt.xlabel('True Stay', fontsize=10)
plt.ylabel('Prediction Stay', fontsize=10)
plt.scatter(truth, prediction)
plt.xlim(-1,30)
plt.ylim(-1,30)

image = plt.show()
display(image)
```

```
import numpy as np

tuples = np.stack((truth,np.round(prediction,1)), axis=-1) ## ROUND TUPLES

frequencies = {}
for x,y in tuples:
    key = (x, y)
    if key in frequencies:
        frequencies[key] += 1
    else:
        frequencies[key] = 1

size = []
x = []
y = []
for key in frequencies.keys():
    size.append(frequencies[key]*10)
    x.append(key[0])
    y.append(key[1])
```

```
plt.figure(figsize=(8, 8), dpi= 300)
plt.xlabel('True Stay', fontsize=10)
plt.ylabel('Prediction Stay', fontsize=10)
plt.scatter(x,y,s=size)
plt.xlim(-1,30)
plt.ylim(-1,30)

image = plt.show()
display(image)
```
